```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('per_capita_income.csv')
#data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
print(data.columns)
```

```python

# display the first 5 rows
data.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income')
plt.scatter(data.Year,data.income,color='red',marker='+')
```

```python
new_data = data.drop('income',axis='columns')
new_data
```

```python
income = data.income
income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_data,income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2003]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
income_pred=reg.predict(new_data)
data['income_prediction']=income_pred
data
```
